```python
# Cell 1: Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential, Model
from keras.layers import LSTM, Dense, Input, Flatten, Reshape, LeakyReLU, BatchNormalization
from keras.optimizers import Adam
from keras.datasets import mnist
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
# Cell 2: Load and preprocess the data
# Load dataframe (replace 'path_to_dataframe.csv' with your dataframe's path)
# url = 'http://apmonitor.com/pds/uploads/Main/auto_trip.zip'
# data = pd.read_csv(url)

df = pd.read_csv('C:/Users/bis1abt/Documents/Projects/TestGAN/Auto_Data_Train.csv')

# Normalize dataframe to [-1, 1]
scaler = MinMaxScaler(feature_range=(-1, 1))
df_normalized = scaler.fit_transform(df)
# Define time steps features for LSTM GAN
time_steps, n_features = df_normalized.shape
# 
print(df_normalized.shape)
print(df_normalized)

#Plot normalized dataset
fig, axs = plt.subplots(1, 3, figsize=(20, 5))
# Plotting data for the first column
axs[0].plot(df_normalized[:, 0], color='r')
axs[0].set_title('Throttle Position (Normalized)')
# Plotting data for the second column
axs[1].plot(df_normalized[:, 1], color='g')
axs[1].set_title('Vehicle Acceleration (Normalized)')
# Plotting data for the third column
axs[2].plot(df_normalized[:, 2], color='b')
axs[2].set_title('Vehicle Speed (Normalized)')
plt.tight_layout()
plt.show()

# time_steps = 28   # using MNIST data which is 28x28
# n_features = 28
# (X_train, _), (_, _) = mnist.load_data()
# X_train = (X_train.astype(np.float32) - 127.5) / 127.5  # Normalize to [-1, 1]
# X_train = np.expand_dims(X_train, axis=-1)  # Expand dimensions for compatibility
# X_train = np.squeeze(X_train, axis=-1)
# # Checking the shape of the data (useful for troubleshooting)
# print(X_train.shape)
```

```python
# Cell 3: Define the generator
def build_generator(time_steps, n_features):
    model = Sequential()
    model.add(LSTM(128, input_shape=(time_steps, n_features)))
    model.add(Dense(256))
    model.add(LeakyReLU(0.2))
    model.add(BatchNormalization(momentum=0.8))
    model.add(Dense(512))
    model.add(LeakyReLU(0.2))
    model.add(BatchNormalization(momentum=0.8))
    model.add(Dense(np.prod((time_steps, n_features)), activation='tanh'))
    model.add(Reshape((time_steps, n_features)))
    return model
```

```python
# Cell 4: Define the discriminator
def build_discriminator(time_steps, n_features):
    model = Sequential()
    model.add(LSTM(128, input_shape=(time_steps, n_features)))
    model.add(Dense(256))
    model.add(LeakyReLU(0.2))
    model.add(Dense(1, activation='sigmoid'))
    return model
```

```python
# Cell 5: Define GAN by combining generator and discriminator
def build_gan(generator, discriminator):
    z = Input(shape=(time_steps, n_features))
    generated_data = generator(z)
    validity = discriminator(generated_data)
    gan = Model(z, validity)
    gan.compile(loss='binary_crossentropy', optimizer=Adam(0.0002, 0.5))
    return gan
```

```python
# # Cell 6: Visualization function for generated samples
def plot_generated_samples(epoch, generated_series, real_series_samples):
    """
    Plot the generated time series samples along with real samples for a given epoch.
    
    Parameters:
    - epoch (int): Current epoch number.
    - generated_series (array-like): Generated samples from the generator.
    - real_series_samples (array-like): Actual real samples from the dataset.
    """
    plt.figure(figsize=(15, 10))
    
    # Plotting generated series
    for i in range(3):  
        plt.subplot(8, 5, i+1)
        plt.plot(generated_series[i, :, 0], label="Generated", color='red')
        plt.title(f"Generated {i+1}")
        # plt.axis('off')
        
    # Plotting real series
    for i in range(3):  
        plt.subplot(8, 5, i+3)
        plt.plot(real_series_samples[i, :, 1], label="Real", color='blue')
        plt.title(f"Real {i+1}")
        # plt.axis('off')
    
    plt.tight_layout()
    plt.suptitle(f"Epoch: {epoch+1}")
    plt.show()
```

```python
# Cell 7: Train the GAN with visualization
def train_gan(generator, discriminator, gan, epochs, batch_size=128, sample_interval=1):
    # Create lists for saving loss values
    d_losses = []
    g_losses = []
    # Create Labels for real (1s) and fake (0s) samples
    valid = np.ones((batch_size, 1))
    fake = np.zeros((batch_size, 1))
    # Main training loop
    for epoch in range(epochs):
        # ---------------------
        #  Train Discriminator
        # ---------------------
        # Select a random batch of real data samples
        idx = np.random.randint(0, df_normalized.shape[0] - time_steps + 1, batch_size)
        sequences = [df_normalized[i:i+time_steps] for i in idx]
        for seq in sequences:
            assert len(seq) == time_steps, f"Found sequence of length {len(seq)}!"
        real_data = np.array(sequences)
        # Generate a random batch of noise samples
        noise = np.random.normal(0, 1, (batch_size, time_steps, n_features))
        # Generate a batch of synthetic data samples
        generated_data = generator.predict(noise)
        # Train the discriminator
        d_loss_real = discriminator.train_on_batch(real_data, valid)
        d_loss_fake = discriminator.train_on_batch(generated_data, fake)
        d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)
        # Generate a batch of noise samples
        noise = np.random.normal(0, 1, (batch_size, time_steps, n_features))
        # Train the generator
        g_loss = gan.train_on_batch(noise, valid)
        # Save and print the losses
        d_losses.append(d_loss[0])
        g_losses.append(g_loss)
        print(f"{epoch}/{epochs} [D loss: {d_loss[0]} | D accuracy: {100 * d_loss[1]}] [G loss: {g_loss}]")
        if epoch % sample_interval == 0:
            plot_generated_samples(epoch, generated_data, real_data)
            print(generated_data.shape)
    
    plt.figure(figsize=(15,5))
    plt.plot(d_losses, label="Discriminator Loss")
    plt.plot(g_losses, label="Generator Loss")
    plt.title("Training Losses")
    plt.legend()
    plt.show()

```

```python
# Cell 8: Initialize and compile the models

discriminator = build_discriminator(time_steps, n_features)
discriminator.compile(loss='binary_crossentropy', optimizer=Adam(0.0002, 0.5), metrics=['accuracy'])
generator = build_generator(time_steps, n_features)
gan = build_gan(generator, discriminator)
discriminator.trainable = False
# View G and D structures for troubleshooting
# print(generator.summary())
# print(discriminator.summary())
```

```python
# Cell 9: Train the GAN
train_gan(generator, discriminator, gan, epochs=5)
```
